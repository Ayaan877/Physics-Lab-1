```python
import numpy as np                  
import matplotlib.pyplot as plt     
%matplotlib inline                   
import scipy as scp  
from scipy.optimize import curve_fit
import statistics as st
```

```python
time = np.array([4.88, 4.84, 4.79, 4.72, 4.81, 4.72, 4.84, 4.88, 4.82, 4.75, 4.81, 4.82, 4.81, 4.78, 4.72, 4.84, 4.75, 4.81, 4.78, 4.88, 4.75])

Mean = np.round(st.mean(time), 3)
Std_Dev = np.round(st.stdev(time), 3)

plt.hist(time, bins=10)
plt.show()

print('Mean = ', Mean)
print('Standard Devation = ', Std_Dev)
```

```python
SM_Axis = np.array([0.3871, 0.7233, 1.000, 1.5273, 2.76596, 5.2028, 9.5388, 19.1914, 30.0611, 39.5294, 43.335, 45.791, 67.6681])
Period = np.array([0.2408, 0.6152, 1, 1.8809, 4.599, 11.862, 29.458, 84.01, 164.79, 248.54, 285.4, 309.88, 557])

def f(x,a,b,c):
    return a*x**2+b*x+c

par, covariance = curve_fit(f, SM_Axis, Period)

p = np.round(par[0],3)
q = np.round(par[1],3)
r = np.round(par[2],3)

Period_trend = p*SM_Axis**2+q*SM_Axis+r

plt.scatter(SM_Axis, Period)
plt.plot(SM_Axis, Period_trend,'--', color="red")

plt.title(r'Variation of Time Period with Semimajor Axis')
plt.xlabel(r'Semimajor Axis (AU)')
plt.ylabel(r'Time Period (Yr)')
plt.show()
```

```python
R_3 = SM_Axis**3
T_2 = Period**2

def f(x,a,b):
    return a*x+b

par, covariance = curve_fit(f, R_3, T_2)

p = np.round(par[0],3)
q = np.round(par[1],3)

T2_trend = p*R_3+q

plt.scatter(R_3, T_2)
plt.plot(R_3,T2_trend,'--', color="red")

plt.text(50000,200000, r'f(x) = '+f'{p}'+'x'+f'{q}', fontsize = 12)

plt.title(r'Variation of Time Period$^2$ with Semimajor Axis$^3$')
plt.xlabel(r'Semimajor Axis$^3$ (AU$^3$)')
plt.ylabel(r'Time Period$^2$ (Yr$^2$)')
plt.show()
```

```python
R_log = np.log(SM_Axis)
T_log = np.log(Period)

def f(x,a,b):
    return a*x+b

par, covariance = curve_fit(f, R_log, T_log)

p = np.round(par[0],3)
q = np.round(par[1],3)

Tlog_trend = p*R_log+q

plt.scatter(R_log, T_log)
plt.plot(R_log,Tlog_trend,'--', color="red")

plt.text(0,4, r'f(x) = '+f'{p}'+'x'+f'{q}', fontsize = 12)

plt.title(r'Variation of ln(Time Period) with ln(Semimajor Axis)')
plt.xlabel(r'ln(Semimajor Axis)')
plt.ylabel(r'ln(Time Period)')
plt.show()
```
