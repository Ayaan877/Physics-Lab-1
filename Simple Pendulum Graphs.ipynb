```python
import numpy as np                  
import matplotlib.pyplot as plt     
%matplotlib inline                   
import scipy as scp  
from scipy.optimize import curve_fit
import statistics as st
```

```python
Length, T1 = np.loadtxt('../Simple Pendulum/L vs T.txt', unpack = True) 
Length, T2 = np.loadtxt('../Simple Pendulum/L vs T^2.txt', unpack = True) 
Mass, T3 = np.loadtxt('../Simple Pendulum/M vs T.txt', unpack = True)
Amp, T4 = np.loadtxt('../Simple Pendulum/Amp vs T.txt', unpack = True)
Amp2, T5 = np.loadtxt('../Simple Pendulum/Amp^2 vs T.txt', unpack = True)
```

```python
def f(x,a,b):
    return a*x+b

par, covariance = curve_fit(f, Length, T1)

p = np.round(par[0],3)
q = np.round(par[1],3)

T1_trend = p*Length+q

plt.scatter(Length,T1)
plt.title('Variation of Time Period with Length')
plt.text(20,1.2, r'f(x) = '+f'{p}'+'x+'+f'{q}', fontsize = 12)

plt.plot(Length,T1_trend,'--',color="blue")
plt.ylim(0.5, 1.5)
plt.xlabel(r'Length (cm)')
plt.ylabel(r'Time (s)')
plt.show()
```

```python
def f(x,a,b):
    return a*x+b

par, covariance = curve_fit(f, Length, T2)

p = np.round(par[0],3)
q = np.round(par[1],3)

T2_trend = p*Length+q

plt.scatter(Length,T2)
plt.title(r'Variation of (Time Period)$^2$ with Length')
plt.text(20,1.4, r'f(x) = '+f'{p}'+'x+'+f'{q}', fontsize = 12)

plt.plot(Length,T2_trend,'--',color="blue")
plt.xlabel(r'Length (cm)')
plt.ylabel(r'Time$^2$ (s$^2$)')
plt.show()
```

```python
def f(x,a,b):
    return a*x+b

par, covariance = curve_fit(f, Mass, T3)

p = np.round(par[0],3)
q = np.round(par[1],3)

T3_trend = p*Mass+q

plt.scatter(Mass,T3)
plt.title('Variation of Time Period with Mass')
plt.text(18,1.15, r'f(x) = '+f'{p}'+'x+'+f'{q}', fontsize = 12)

plt.plot(Mass,T3_trend,'--',color="blue")
plt.ylim(0.5, 1.5)
plt.xlabel(r'Mass (g)')
plt.ylabel(r'Time (s)')
plt.show()
```
