```python
import numpy as np                  
import matplotlib.pyplot as plt     
%matplotlib inline                   
import scipy as scp  
from scipy.optimize import curve_fit
import statistics as st
```

```python
time, x_val, y_val = np.loadtxt('../Simple Pendulum - Video Tracking/Video Tracking 3 (damping).txt', unpack = True, max_rows=10000)
```

```python
#Planar Pendulum Graph

A = 0.0369
T = 1.261
p = -2.778
k = 0.0077
d = -0.0083

plt.plot(time, A*np.exp(d*time)*np.sin((2*np.pi*time)/T + p)+k , color = "red", linewidth=1)

plt.scatter(time, x_val, s=10)
plt.xlabel(r'Time [s]')
plt.ylabel(r'Amplitude [m]')
plt.title('x-axis amplitude vs time')
#plt.text(70, 0.035, r'$A(t)=A\sin(\frac{2\pi t}{T}+\phi) e^{dt}+k$', fontsize = 12)
plt.show
print(len(x_val))
```

```python
mx = []
mn = []
t1 = []
t2 = []

for i in range(1, 9999):
 
        if(x_val[i-1] > x_val[i] < x_val[i + 1] and x_val[i]<0):
            mn.append(x_val[i])
            t2.append(time[i])
 
        elif(x_val[i-1] < x_val[i] > x_val[i + 1] and x_val[i]>0):
            mx.append(x_val[i])
            t1.append(time[i])
```

```python
mx = np.array(mx)
t1 = np.array(t1)

p = np.polyfit(t1, np.log(mx), 1)
print(p)

m = -0.006
c = -3.136

plt.scatter(t1, np.log(mx))
plt.plot(t1, m*t1+c, color = "red")
plt.xlabel(r'Time [s]')
plt.ylabel(r'ln(Max Amplitude) [$\ln(A_{m})$]')
plt.title('log of maximum amplitude vs time')
plt.text(50, -3.3, r'f(x) = '+f'{m}'+'x'+f'{c}', fontsize = 12)
plt.show()
```

```python
t, x_t, y_t = np.loadtxt('../Simple Pendulum - Video Tracking/Elliptical Pendulum Data 1.txt', unpack = True)
```
